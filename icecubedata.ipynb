```python
import matplotlib.pyplot as plt
import numpy as np

# Ruta al archivo de datos
file_path = '/Aeff_IC86b.txt'

# Listas para almacenar los datos
log10_Emin = []
log10_Emax = []
Aeff = []

# Leer el archivo lnea por lnea
with open(file_path, 'r') as file:
    for line in file:
        if line.startswith('#') or line.strip() == '':
            continue  # Saltar lneas vacas y comentarios
        parts = line.split()
        try:
            log10_Emin.append(float(parts[0]))
            log10_Emax.append(float(parts[1]))
            Aeff.append(float(parts[2]))
        except ValueError:
            continue  # Saltar lneas que no se pueden convertir a float

# Convertir las listas a arrays de numpy
log10_Emin = np.array(log10_Emin)
log10_Emax = np.array(log10_Emax)
Aeff = np.array(Aeff)

# Calcular los valores medios de los intervalos de energa
log10_Emean = (log10_Emin + log10_Emax) / 2

# Crear la grfica
plt.figure(figsize=(10, 6))
plt.plot(10**log10_Emean, Aeff, marker='o', linestyle='-', color='b')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Energa de los Neutrinos (GeV)')
plt.ylabel('rea Efectiva (m)')
plt.title('rea Efectiva del Detector IceCube en funcin de la Energa de los Neutrinos')
plt.grid(True)
plt.show()
```
